```python
import pandas as  pd
from assets.data.adwotd.matlab_AD_3axisVibrationData.get_data import get_data
train_x, train_y, test_x, test_y = get_data(train_nornal_only=True, test_normal_only=False)
train_x = pd.DataFrame(train_x)
time_index = pd.to_datetime(train_x.index, unit='h', origin='2022-01-01')  # 这里使用了一个起始日期 '2022-01-01'，你可以根据需要更改
train_x.index = time_index


test_x = pd.DataFrame(test_x)
time_index = pd.to_datetime(test_x.index, unit='h', origin='2022-01-01')  # 这里使用了一个起始日期 '2022-01-01'，你可以根据需要更改
test_x.index = time_index
train_x
```

### adwotd

```python
from adwotd.models.pca import PCA
from adwotd.utils.evaluate import acc_balacc_recall_f1, _confusion_matrix
from adwotd.utils.save_load_model import jolib_save
from adwotd.utils.visualize import scatter_plot, probability_distribution_plot,probability_distribution_plot_twoClass_realLabel
# =========================== initialize and train the model ===========================
clf_name = 'PCA'
clf = PCA(contamination=0.1)
clf.fit(train_x)

# =========================== result on training data ===========================
train_y_scores = clf.decision_scores_  # train scores
th = clf.threshold_
train_y_pred = clf.labels_  # train labels (0 or 1)

print('--------- train ---------')
print('score:', train_y_scores)
print('threshold:', th)
print('pred:',train_y_pred)

# roc_and_prn(train_y, train_y_scores)
acc_balacc_recall_f1(train_y, train_y_pred)
_confusion_matrix(train_y, train_y_pred)
scatter_plot(train_y_scores, th)
probability_distribution_plot(train_y_scores, threshold=th)

# =========================== result on testing data ===========================
# get the prediction on the test data
test_y_scores = clf.decision_function(test_x)  # test scores
test_y_pred = clf.predict(test_x)  # tset labels (0 or 1)
test_y_proba, test_y_confidence = clf.predict_proba(test_x, return_confidence=True)

print('--------- test ---------')
print('score:', test_y_scores)
print('threshold:', th)
print('pred:', test_y_pred)
print('confidence:', test_y_confidence)
print('proba:', test_y_proba)

# roc_and_prn(test_y, test_y_scores)
acc_balacc_recall_f1(test_y, test_y_pred)
_confusion_matrix(test_y, test_y_pred)


scatter_plot(test_y_scores, th)
probability_distribution_plot_twoClass_realLabel(train_y_scores, test_y_scores, th)


# =========================== model save ===========================
jolib_save(clf, clf_name + '_matlab_AD_3axisVibrationData.joblib')

```

## generate data

```python
import numpy as np
import pandas as pd
from datetime import datetime

# 生成日期范围
start_date = datetime(2022, 10, 1)
end_date = datetime(2023, 12, 31)
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# 生成每天的随机数据
np.random.seed(42)
data_values = np.random.normal(size=len(date_range))
# 在2023.09.26这一天发生Pattern Change异常
change_date_pattern = datetime(2023, 9, 26)
change_index_pattern = date_range.get_loc(change_date_pattern)
data_values[change_index_pattern:] += np.random.normal(0, 5, len(data_values) - change_index_pattern)  # 添加随机扰动

# 在中断位置插入Level Shift异常（例如，2023.07.15）
shift_date = datetime(2023, 7, 15)
shift_index = date_range.get_loc(shift_date)
data_values[:shift_index] -= 3  # 增加一个恒定的偏移量

# 将特定日期的值更改为7.5
data_values[date_range == datetime(2023, 3, 9)] = 7.5

# 创建数据框
LevelShift_and_PatternChange = pd.DataFrame({'Date': date_range, 'LevelShift_and_PatternChange': data_values})
# 将 "Date" 列设置为索引
LevelShift_and_PatternChange.set_index('Date', inplace=True)

# 转换索引为 DatetimeIndex 对象
LevelShift_and_PatternChange.index = pd.to_datetime(LevelShift_and_PatternChange.index)
LevelShift_and_PatternChange
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
plt.plot(LevelShift_and_PatternChange)
```

```python
start_date_range = '2022-10-01'
end_date_range = '2023-02-15'
train_x = LevelShift_and_PatternChange.loc[start_date_range:end_date_range]

plt.figure(figsize=(15, 5))
plt.plot(train_x)
```

```python
test_x = LevelShift_and_PatternChange.loc[end_date_range:]
test_x
plt.figure(figsize=(15, 5))
plt.plot(test_x)
```
